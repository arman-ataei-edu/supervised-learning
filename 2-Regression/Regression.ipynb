```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot

import seaborn as sns

# make regression 
from sklearn.datasets import make_regression

from sklearn.model_selection import train_test_split
```

```python
X,Y = make_regression(400, 1, n_informative=1, noise=60)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2)
# The parameters for subplot are: 
# number of rows, number of columns, 
# and which subplot you're currently on. 
print(X.shape, Y.shape)
plt.subplot(1,3,1)
plt.scatter(X,Y)
plt.title("Made dataset")
# plt.show()

plt.subplot(1,3,2)
plt.scatter(X_train, Y_train)
plt.title("Train Set")
# plt.show()


plt.subplot(1,3,3)
plt.scatter(X_test, Y_test)
plt.title("Test Set")
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)

# y_pred = lr.predict(X_test)
score = lr.score(X_test, Y_test)
print(score)
```

```python
_x_range = np.arange(-3, 3, .1)
_x_range_re = np.reshape(_x_range, (-1, 1))
y_pred = lr.predict(_x_range_re)

plt.scatter(X,Y, color="red")
plt.plot(_x_range_re, y_pred)
plt.title("Prediction")
```
