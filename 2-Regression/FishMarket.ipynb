```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split


```

### preprocessing data

```python
data = pd.read_csv("./Fish.csv").set_axis([i for i in range(7)], axis=1)
# converting the categorical data into numericals
one_hot = pd.get_dummies(data[0])
raw_data = data.drop(0, axis=1)
data = raw_data.join(one_hot)

# Normalizing the data
cols = [i for i in range(1,7)]
min_max_scaler = MinMaxScaler()
x = data[cols].values
x_scaled = min_max_scaler.fit_transform(x)

normalized = pd.DataFrame(x_scaled, columns=cols, index=data.index)
data[cols] = normalized
data
```

### Choosing Target variable and Spliting the dataset

```python
Y = np.array(data.pop(1))
X = np.array(data)

print(X.shape, Y.shape)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.2)
print(X_test.shape, Y_test.shape)
```

### Init and Training the Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
# using linear algebra (normal equation and inverse/pesudo inverse)
lr = LinearRegression()
lr.fit(X_train, Y_train)

score = lr.score(X_test, Y_test)
print(score)

sampls_5_x = X_test[:5,:]
sampls_5_y = Y_test[:5]

print(sampls_5_y, lr.predict(sampls_5_x), sep="\n")
```
