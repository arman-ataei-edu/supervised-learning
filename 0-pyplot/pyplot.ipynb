```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_regression
from scipy.interpolate import griddata

# Generate data
X, Y = make_regression(n_samples=400, n_features=2, n_informative=2, noise=2.5)

# Extract the two features
X1, X2 = X[:, 0], X[:, 1]

# Create grid values
grid_x, grid_y = np.mgrid[np.min(X1):np.max(X1):100j, np.min(X2):np.max(X2):100j]

# Interpolate data
grid_z = griddata((X1, X2), Y, (grid_x, grid_y), method='cubic')

# Plot the surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(grid_x, grid_y, grid_z, cmap='viridis')

# Adding labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('3D Surface Plot')

plt.show()
# copilot response
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

x = np.linspace(0, 10, 6)  # [0, 2,..,10] : 6 distinct values
y = np.linspace(0, 20, 5)  # [0, 5,..,20] : 5 distinct values
z = np.linspace(0, 100, 30)  # 6 * 5 = 30 values, 1 for each possible combination of (x,y)

X, Y = np.meshgrid(x, y)

Z = np.reshape(z, X.shape)  # Z.shape must be equal to X.shape = Y.shape

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()

# https://stackoverflow.com/questions/9170838/surface-plots-in-matplotlib
```
